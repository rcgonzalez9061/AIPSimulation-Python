```python
from twitterscraper import query_tweets
import pandas as pd
import datetime
import numpy as np
```

```python
being_date = datetime.date(2019, 12, 30)
end_date = datetime.date(2020, 2, 19)

lang="english"
```

```python
tweets = query_tweets("#coronavirus", begindate=being_date, enddate=end_date, lang=lang)
```

```python
df = pd.DataFrame(t.__dict__ for t in tweets)
df
```

```python
df.columns
```

```python
df[['likes', 'retweets', 'replies', "text", "screen_name", "tweet_url"]].sort_values('likes', ascending=False).head(35)
```

```python
df['hashtags'].explode().str.lower().value_counts().head(15)
```

```python
wuhan_tweets = query_tweets("#wuhan", begindate=being_date, enddate=end_date, lang=lang)
```

```python
wuhan_df = pd.DataFrame(t.__dict__ for t in wuhan_tweets)
wuhan_df[['likes', 'retweets', 'replies', "text", "screen_name", "tweet_url"]].sort_values('likes', ascending=False).head(35)
```

```python
df.to_csv("corona")
```
